This task identifies missing values, data type issues, duplicate records, and inconsistent entries in the 2022 HDI dataset.

### Task 6: HDI Category Classification

This task classifies countries into official UNDP HDI categories based on their HDI values for the year 2022.


```python
# Function to classify HDI category
def classify_hdi(hdi_value):
    if hdi_value < 0.550:
        return 'Low'
    elif 0.550 <= hdi_value <= 0.699:
        return 'Medium'
    elif 0.700 <= hdi_value <= 0.799:
        return 'High'
    else:
        return 'Very High'
```

```python
# Apply HDI category classification
hdi_2022_df['HDI Category'] = hdi_2022_df['hdi'].apply(classify_hdi)

# Display sample rows to verify
hdi_2022_df[['country', 'hdi', 'HDI Category']].head()
```

```python
# Save the final dataframe with HDI category
hdi_2022_df.to_csv('HDI_category_added.csv', index=False)
```

**Interpretation:**

The HDI Category column classifies countries into four development groups as defined by the UNDP.
This classification helps in comparing development levels across countries and supports further regional and trend-based analysis.

### Task 2: Composite Development Score

A composite score is calculated using life expectancy and GNI per capita to compare development levels across South Asian countries.


```python
# Group by country and take mean Composite Score
south_asia_df_copy = south_asia_df.copy()
south_asia_df_copy['Composite Score'] = south_asia_df_copy['life_expectancy'] * south_asia_df_copy['gross_inc_percap']

composite_grouped = (
    south_asia_df_copy
    .groupby('country', as_index=False)['Composite Score']
    .mean()
)

# Sort by Composite Score
composite_grouped = composite_grouped.sort_values(
    by='Composite Score', ascending=False
)

composite_grouped
```

```python
# Select top 5 countries
top_5_countries = (
    composite_grouped
    .head(5)
    .reset_index(drop=True)
)

# Plot using matplotlib
plt.figure(figsize=(8, 5))
plt.barh(
    top_5_countries['country'],
    top_5_countries['Composite Score']
)

plt.xlabel('Composite Score')
plt.ylabel('Country')
plt.title('Top 5 South Asian Countries by Composite Score')

# Highest score on top
plt.gca().invert_yaxis()

plt.show()
```

**Interpretation:**

Countries with higher life expectancy and income levels rank higher in the composite score.
Differences between composite score ranking and HDI ranking highlight variations in development dimensions.


### Task 3: Outlier Detection using IQR

Outliers in HDI and GNI per capita are detected using the 1.5×IQR rule.


```python
# IQR for HDI
Q1_hdi = south_asia_df['hdi'].quantile(0.25)
Q3_hdi = south_asia_df['hdi'].quantile(0.75)
IQR_hdi = Q3_hdi - Q1_hdi

# IQR for GNI
Q1_gni = south_asia_df['gross_inc_percap'].quantile(0.25)
Q3_gni = south_asia_df['gross_inc_percap'].quantile(0.75)
IQR_gni = Q3_gni - Q1_gni

# Identify outliers
outliers = south_asia_df[
    (south_asia_df['hdi'] < Q1_hdi - 1.5 * IQR_hdi) |
    (south_asia_df['hdi'] > Q3_hdi + 1.5 * IQR_hdi) |
    (south_asia_df['gross_inc_percap'] < Q1_gni - 1.5 * IQR_gni) |
    (south_asia_df['gross_inc_percap'] > Q3_gni + 1.5 * IQR_gni)
]

outliers[['country', 'hdi', 'gross_inc_percap']]
```

```python
# Scatter plot showing outliers
plt.figure(figsize=(7, 5))

# Normal data points
plt.scatter(
    south_asia_df['gross_inc_percap'],
    south_asia_df['hdi'],
    label='Normal',
    alpha=0.7
)

# Outliers highlighted
plt.scatter(
    outliers['gross_inc_percap'],
    outliers['hdi'],
    color='red',
    label='Outliers'
)

plt.xlabel('Gross National Income per Capita')
plt.ylabel('HDI')
plt.title('Outlier Detection: HDI vs GNI per Capita (South Asia)')
plt.legend()
plt.show()
```

**Interpretation:**

Outliers represent countries whose income or HDI significantly deviates from the regional pattern.
Such deviations may arise due to economic inequality, conflict, or policy differences.

```python
def detect_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    return df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]

out_sa = detect_outliers(south_asia_2020_2022, 'hdi')
out_me = detect_outliers(middle_east_2020_2022, 'hdi')

out_sa[['country','hdi']], out_me[['country','hdi']]
```

**Interpretation:**

Outliers highlight countries with unusually high or low HDI values compared to regional norms.
These may reflect conflict, economic concentration, or exceptional development policies.
